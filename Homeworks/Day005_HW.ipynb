```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

# --- ANS --- #

import requests
import json

path = r'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
res = requests.get(path).text

resj = json.loads(res)

print(type(resj))
print('-----------------\n')

'''
for i in range(0, len(resj)):
    print(type(resj[i]))
    print(resj[i])
    print('-----------------\n')
'''

    
print('######### --- ANS 1 --- #############')
print('ä¸€æ¬¡è«‹æ±‚å…±è¨ˆå›å‚³', len(resj), 'ç­†è³‡æ–™')
print('-----------------\n')


print('ä»¥å…¶ä¸­ä¸€ç­†è³‡æ–™ç‚ºä¾‹ï¼Œå…¶ä¸­åŒ…å«æ¬„ä½å¦‚ä¸‹ï¼š')
for j, k in  resj[1].items():
    print('æ¬„ä½ï¼š', j, '--', 'ex:', k)
print('-----------------\nå…±è¨ˆï¼š', len(resj[1]))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

d1 = {'æ¨™é¡Œ': 'title', 'è²¼æ–‡æ™‚é–“': 'createdAt', 'ç•™è¨€äººæ•¸': 'commentCount', 'æŒ‰è®šäººæ•¸': 'likeCount'}
print('######### --- ANS 2 --- #############')
print('ä»¥ä¸€æ¬¡è«‹æ±‚å…±è¨ˆå›å‚³', len(resj), 'ç­†è³‡æ–™ç‚ºä¾‹\n')

for n in range(0, len(resj)):
    print('~ç¬¬%dç­†è³‡æ–™åˆ†æå¦‚ä¸‹ï¼š' %(n+1))
    for o in d1.keys():
        print('\t', o, ' : ', resj[n][d1[o]])
    print('==================')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

print('######### --- ANS 3 --- #############')
print('(ä»¥ ç•™è¨€äººæ•¸ ä½œç‚ºç†±é–€æ–‡ç« ä¹‹ä¾æ“š)')
print('ä»¥ä¸€æ¬¡è«‹æ±‚å…±è¨ˆå›å‚³', len(resj), 'ç­†è³‡æ–™ç‚ºä¾‹\n')

comc = []
likc = []
dcom = {}
dlik = {}

for n in range(0, len(resj)):
    comc.append(resj[n]['commentCount'])
    likc.append(resj[n]['likeCount'])
    dcom[resj[n]['commentCount']] = n
    dlik[resj[n]['commentCount']] = n

    
#print(comc)
#print(likc)

#print(sorted(comc))
sorcomc = sorted(comc)

if len(sorcomc) % 2 == 0:
    ma = len(sorcomc)/2
    mi = len(sorcomc)/2 - 1
    #print(ma, mi)
    madi = (sorcomc[int(ma)] + sorcomc[int(mi)])/2
madi

zc = 0
zl = 0
znc = 0
znl = 0
for x, y in enumerate(comc):
    #print(x, y)
    if y > madi:
        zc += y
        zl += likc[x]
    else:
        znc += y
        znl += likc[x]

print('==================')
print('ç†±é–€æ–‡ç« å…±è¨ˆ%dç­†ï¼š' %(len(resj)/2), '\n\tå¹³å‡ç•™è¨€äººæ•¸ï¼š%d' %(zc/(len(resj)/2)), '\n\tå¹³å‡æŒ‰è®šäººæ•¸ï¼š%d' %(zl/(len(resj)/2)))
print('\néç†±é–€æ–‡ç« å…±è¨ˆ%dç­†ï¼š' %(len(resj)/2), '\n\tå¹³å‡ç•™è¨€äººæ•¸ï¼š%d' %(znc/(len(resj)/2)), '\n\tå¹³å‡æŒ‰è®šäººæ•¸ï¼š%d' %(znl/(len(resj)/2)))
```
