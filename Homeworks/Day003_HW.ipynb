### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異

```python
import urllib.request
import zipfile
import os, sys
import xmltodict
import pandas as pd 

res = r'http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314'

try:
    os.makedirs('./xmldata', exist_ok = True)
except:
    print("ERROR!!  -pls recheck- no.1")
    

urllib.request.urlretrieve(res, './xmldata/sample.zip')

if os.path.exists('./xmldata/sample.zip'):
    zipfile.ZipFile('./xmldata/sample.zip').extractall('./xmldata')
    #print(fxml, '\n ---------- \n')

try:
    dirs = os.listdir('./xmldata')
except:
    print("ERROR!!  -pls recheck- no.2")

ii = 0
for i in dirs:
    print(i)
    ii += 1
print('\n---------total files-----------\n|--> : ', ii)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
try:
    dirs = os.listdir('./xmldata')
except:
    print("ERROR!!  -pls recheck- no.2")
sp = []

for file in dirs:
    filep = os.path.join('./xmldata/', file)
    if "xml" in os.path.splitext(filep)[1]:
        #print(filep)
        with open(filep, 'r', encoding = 'UTF8') as ff:
            fxml = ff.read()
        FD = xmltodict.parse(fxml)

        if ('高雄市' in FD['cwbopendata']['dataset']['locations']['locationsName']):
            sp.append(filep)
            next;
        else:
            for il in FD['cwbopendata']['dataset']['locations']['location']:
                if '高雄市' in il['locationName']:
                    sp.append(filep)
                    next;
                
    else:
        print('\n\n not be included in the .xml file!\t\t', filep, '\n\n')

print(sp)
```

```python
# 1. 請問高雄市有多少地區有溫度資料？
city = '高雄市'

for xp in sp:

    with open(xp, 'r', encoding = 'utf-8') as fxp:
        ffxp = fxp.read()
    FXP = xmltodict.parse(ffxp)
    
    title = FXP['cwbopendata']['dataset']['datasetInfo']['datasetDescription']

    N_city = FXP['cwbopendata']['dataset']['locations']['locationsName']
    
    if (N_city == city):
        count = 0
        Tarea = []
        for loc in FXP['cwbopendata']['dataset']['locations']['location']:
            if (loc['locationName'] != city) and ('溫度' in loc['weatherElement'][0]['description']):
                count += 1
                Tarea.append(loc['locationName'])
                
        print('檔案名稱：', xp)
        print(title, ' -- ', city)
        print('\n## 含有(平均)溫度資料之區域個數 ## ：', count)
        print('\n區域名稱：', Tarea, '\n\n')
        
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for xp in sp:

    with open(xp, 'r', encoding = 'utf-8') as fxp:
        ffxp = fxp.read()
    FXP = xmltodict.parse(ffxp)
    
    title = FXP['cwbopendata']['dataset']['datasetInfo']['datasetDescription']

    N_city = FXP['cwbopendata']['dataset']['locations']['locationsName']

    if (N_city == city):
        count = 0
        Tarea = []
        tti = []
        sti = []
        eti = []
        ttt = []
        for loc in FXP['cwbopendata']['dataset']['locations']['location']:
            if ('溫度' == loc['weatherElement'][0]['description']):
                count += 1
                Tarea.append(loc['locationName'])
                tti.append(loc['weatherElement'][0]['time'][0]['dataTime'])
                ttt.append((loc['weatherElement'][0]['time'][0]['elementValue']['value'], loc['weatherElement'][0]['time'][0]['elementValue']['measures']))

            elif ('平均溫度' in loc['weatherElement'][0]['description']):
                count += 1
                Tarea.append(loc['locationName'])
                ttt.append((loc['weatherElement'][0]['time'][0]['elementValue']['value'], loc['weatherElement'][0]['time'][0]['elementValue']['measures']))
                sti.append(loc['weatherElement'][0]['time'][0]['startTime'])
                eti.append(loc['weatherElement'][0]['time'][0]['endTime'])

        print('\n#######   ########   #########   ########   ########')        
        print('檔案名稱：', xp)
        print(title, ' -- ', city)
        print('\n## 含有溫度資料之區域個數 ## ：', count)
        print('\n區域名稱：', Tarea, '\n\n')
        print('==================== 各區記錄如下 ====================')
        for i in range(0, len(Tarea)):
            if len(tti) != 0:
                print('區域：', Tarea[i])
                print('溫度 -- 第一筆記錄時間：', tti[i])
                print('偵測溫度：', ttt[i][0], ttt[i][1])
                print('-----------')
            elif (len(sti) != 0) and (len(eti) != 0):
                print('區域：', Tarea[i])
                print('平均溫度 -- 第一筆記錄開始時間：', sti[i])
                print('平均溫度 -- 第一筆記錄結束時間：', eti[i])
                print('偵測溫度：', ttt[i][0], ttt[i][1])
                print('-----------')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
for xp in sp:

    with open(xp, 'r', encoding = 'utf-8') as fxp:
        ffxp = fxp.read()
    FXP = xmltodict.parse(ffxp)
    
    title = FXP['cwbopendata']['dataset']['datasetInfo']['datasetDescription']
    N_city = FXP['cwbopendata']['dataset']['locations']['locationsName']

    if (N_city == city):
        count = 0
        Tarea = []
        tti = []
        sti = []
        eti = []
        ttt = []
        loc1 = FXP['cwbopendata']['dataset']['locations']['location'][0]
        
        for loct in loc1['weatherElement'][0]['time']:
            if ('溫度' == loc1['weatherElement'][0]['description']):
                count += 1
                Tarea.append(loc1['locationName'])
                tti.append(loct['dataTime'])
                ttt.append((loct['elementValue']['value'], loct['elementValue']['measures']))

            elif ('平均溫度' in loc1['weatherElement'][0]['description']):
                count += 1
                Tarea.append(loc1['locationName'])
                ttt.append((loct['elementValue']['value'], loct['elementValue']['measures']))
                sti.append(loct['startTime'])
                eti.append(loct['endTime'])

        print('\n#######   ########   #########   ########   ########')        
        print('檔案名稱：', xp)
        print(title, ' -- ', city)
        
        print('\n區域名稱：', set(tuple(Tarea)), '\n\n')
        print('\n## 該區所含之溫度資料 ## ：', count)
        print('==================== 各記錄如下 ====================')
        for i in range(0, len(Tarea)):
            if len(tti) != 0:
                print('區域：', Tarea[i])
                print('溫度 -- 第', i+1, '筆記錄時間：', tti[i])
                print('偵測溫度：', ttt[i][0], ttt[i][1])
                print('-----------')
            elif (len(sti) != 0) and (len(eti) != 0):
                print('區域：', Tarea[i])
                print('平均溫度 -- 第', i+1, '筆記錄開始時間：', sti[i])
                print('平均溫度 -- 第', i+1, '筆記錄結束時間：', eti[i])
                print('偵測溫度：', ttt[i][0], ttt[i][1])
                print('-----------')
```
